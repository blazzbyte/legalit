# Analysis of the Documents before proccesing

```python
import pickle

# Load the list from the file
with open('documents.pkl', 'rb') as file:
    documents = pickle.load(file)
```

```python
documents[0]
```

```python
from json import dumps
print(dumps([{'content': doc.text, 'metadata': doc.metadata} for doc in documents], indent=2))
```

```python
len(documents)
```

```python
page_1 = [{'content': doc.text, 'metadata': doc.metadata} for doc in documents if doc.metadata['page_number']==1]
print(dumps(page_1, indent=2))
```
